```python
import pandas as pd
import os
```

```python
data = pd.read_csv("../../../nature_cancer_dataset/single_cell_data.csv")
```

```python
for i, x in data.groupby('ImageNumber'):
    p = os.path.join('../../../output/nature_cancer_cell_positions/', "patient{}_cell_positions.csv".format(i))
    x = x[['Location_Center_X', 'Location_Center_Y', 'description']]
    x.columns = ['x', 'y', 'cell_type']
    x['size_x'] = int(x['x'].max())
    x['size_y'] = int(x['y'].max())
    x.to_csv(p, header=True, index=False, sep=',')
```
